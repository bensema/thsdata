# Tick数据

```python
import pytz
import datetime
from thsdata import Quote


tz = pytz.timezone('Asia/Shanghai')

quote = Quote()

start_dt = tz.localize(datetime.datetime(2024, 12, 31, 9, 25, 0))
end_dt = tz.localize(datetime.datetime(2024, 12, 31, 15, 0, 0))

quote.transaction_history("USHA600519",start_dt)
```

# 完整案例


1.   创建连接
2.   查询数据
3.   断开连接(不用了断开是个好习惯)




```python
from thsdata import Quote


def main():
    # 初始化
    quote = Quote()

    try:
        # quote.connect()
        data = quote.stock_codes()
        print(data)

    except Exception as e:
        print("An error occurred:", e)

    finally:
        # 断开连接
        quote.disconnect()
        print("Disconnected from the server.")


if __name__ == "__main__":
    main()
```
